#### Importing packages and initializing variables

```python
import spotipy
import pandas as pd
from spotipy.oauth2 import SpotifyClientCredentials
SPOTIPY_CLIENT_ID='e18fafeb60a949d2a9b7d1efccabe69a'
SPOTIPY_CLIENT_SECRET='739bbbed49864382a64a64ccd64ecdcc'
auth_manager = SpotifyClientCredentials(client_id=SPOTIPY_CLIENT_ID, client_secret=SPOTIPY_CLIENT_SECRET)
sp = spotipy.Spotify(auth_manager=auth_manager)
sp.me
```

#### Retrieving and storing all playlist information to a variable for access

```python
# Passing the playlist id with all tracks to the Spotipy playlist method and storing it in variable "results"
results = sp.playlist("spotify:playlist:2LdSIU9cmAiAYq3ivWfv0d", additional_types=('track', ))
```

#### Going through the list of data and storing information into dataframe 

```python
# Creating a list of indexes for each number of tracks in the playlist for traversal
num_tracks = list(range(0,len(results["tracks"]["items"])))

# Initializing lists to represent columns to store the values while going through the playlist information
title = []
features = []

# Loop to traverse through the list of tracks
for index in num_tracks:

    # Appending each title to the Title column
    title.append(results["tracks"]["items"][index]["track"]["name"])

    # Finding the current track id in order to pass it though the Spotipy "audio_features" method
    current_track_id = results["tracks"]["items"][index]["track"]["id"]

    # Retrieving the audio features for the current track using the track id and appending it to the Features column
    features.append(sp.audio_features(current_track_id)[0])

# Creating a dataframe with the columns and the information collected
df = pd.DataFrame({"Title": title, "Audio Features": features})

# Splitting the audio features into their own columns and appending to dataframe above
df = pd.concat([df, df["Audio Features"].apply(pd.Series)], axis=1)
df

```

#### Exporting dataframe as CSV file

```python
df.to_csv('spotify_audio_features.csv', index=False)
```
